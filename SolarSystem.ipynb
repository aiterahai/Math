```python
# data : https://nssdc.gsfc.nasa.gov/planetary/factsheet/
# unit : 10^6km

distance = [[46.0, 69.8],
            [107.5, 108.9],
            [147.1, 152.1],
            [206.7, 249.3],
            [740.6, 816.4],
            [1357.6, 1506.5],
            [2732.7, 3001.4],
            [4471.1, 4558.9]]
```

```python
# 1au = 149597870700m

au = 149597870700
```

```python
#10^6km -> au
distance = [[round(j * (10 ** 9) / au, 3) for j in i] for i in distance]
```

```python
distance = distance
distance
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(12,12))
plt.grid(linestyle='-.')

for m, M in distance:
    
    a = (m + M) / 2
    
    c = a - m
    
    b = (a**2 - c**2) ** 0.5
    
    x = np.linspace(-a, a, 10000)
    
    y = (b ** 2 - ((b ** 2) * ((x) ** 2)) / (a ** 2)) ** 0.5
    
    print(round(b, 6), round(a, 6))
    
    plt.plot(x, y, c='k')
    plt.plot(x, -y, c='k')

plt.show()
```
